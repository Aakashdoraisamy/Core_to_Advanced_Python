## Updating dataframe

### how to update the dropna records by default 
### dataframe.dropna(inplace=True)

```python
df = df.dropna()
df
```

```python
import numpy as np
df = pd.DataFrame({'col1':[1,2,3,np.nan,4,5],'col2':[11,23,None,None,100,20]})

df.dropna(inplace=True) # this is another way to update the dataframe
df 
```

```python
import numpy as np
df = pd.DataFrame({'col1':[1,2,3,np.nan,4,5],'col2':[11,23,None,None,100,20]})

df.dropna(subset=['col1','col2'],how='all') 
df
# this is 3rd way to update the dataframe if from the entire rows have null it removes 
# if from the any one coloumn have values then it won't reomve if we want to remove if any one have null value than we use 'any'
# by default 'any' or forcefully need to give all based on the scenarios and column importance we will give 'any' or 'all'
```
