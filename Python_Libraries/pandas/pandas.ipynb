## fillna methods

## 'ffill' --> forward fill
## 'bfill' --> backward fill

## forward fill

```python
print('DataFrame....')
print('Before Forward Fill.....')
print(df)
print('After Forward Fill....')
print(df.fillna(method='ffill')) # it fill the null with previous value
```

## when we have null in first index

```python
import numpy as np
df = pd.DataFrame({'col1':[None,2,3,np.nan,4,5],'col2':[11,23,None,None,100,20]})
print('DataFrame....')
print('Before Forward Fill.....')
print(df)
print('After Forward Fill....')
print(df.fillna(method='ffill')) # after using forward fill it returns NaN only because in previous there is no previous values so NaN
```

# backward fill

## 1st Scenario

```python
import numpy as np
df = pd.DataFrame({'col1':[None,2,3,np.nan,4,None],'col2':[11,23,None,None,100,20]})
print('DataFrame')
print('Before Backward Fill.....')
print(df)
print('After Backward Fill....')
print(df.fillna(method='bfill')) # similarly for backward if there is no backward then it will be NaN
```

## 2nd scenario for backward fill

```python
import numpy as np
df = pd.DataFrame({'col1':[None,2,3,np.nan,None,None],'col2':[11,23,None,None,100,20]})
print('DataFrame')
print('Before Backward Fill.....')
print(df)
print('After Backward Fill....')
print(df.fillna(method='bfill'))
```

```python
import numpy as np
df = pd.DataFrame({'col1':[None,2,3,np.nan,None,None],'col2':[11,23,None,None,100,20]})
df
```
