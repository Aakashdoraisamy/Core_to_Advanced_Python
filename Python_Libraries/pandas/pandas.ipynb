# Pandas

## importing pandas library

```python
import pandas as pd
```

## Basic Series creation

```python
series = pd.Series([1,2,3,4,5],index=['A','B','C','D','E'])
series
```

```python
series = pd.Series((1,2,3,4,5),index=['A','B','C','D','E'])
series
```

## Set is unordered collection so not possible

```python
series = pd.Series({1,2,3,4,5},index=['A','B','C','D','E'])
series
```

```python
series = pd.Series({'A':1,'B':2,'C':3,'D':4,'E':5})
series
```

```python
pd.Series({"Name":['AA','BB'],"Name1":['AA']})
```

## DataFrame Creation:-

## Syntax:-
## pd.DataFrame(dict)

```python
data = pd.DataFrame({'Name':['Batman','Captain America'],'Role':['Leader','Captain']})
data
```

```python
data = pd.DataFrame({'Name':['David Warner','Kane Williamson','Yuvaraj Singh','Shikwar Dhawan',
                             'Aiden Markaram','Pat Cummins','Henrich Klassen','Abishek Sharma',
                             'Nithish Kumar','Bhuvaneshwar Kumar','Travis Head'],
                     'Country':['Aus','NZ','Ind','Ind','SA','Aus','SA','Ind','Ind','Ind','Aus']})
data
```

```python
data = pd.DataFrame([[1,2],[4,5]],columns=['col1','col2'])
data
```
